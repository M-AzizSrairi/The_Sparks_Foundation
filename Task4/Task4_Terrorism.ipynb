```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Import the dataset
df = pd.read_excel("/content/globalterrorismdb_0718dist.xlsx")

df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
# number of incidents per year
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='iyear',palette='muted')
plt.title('Number Of Terrorist Activities Each Year')
plt.xticks(rotation=90)
```

```python
import seaborn as sns
# number of incidents per month
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='imonth',palette='muted')
plt.title('Number Of Terrorist Activities Each Year')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='country_txt', palette='muted',
              order=df['country_txt'].value_counts().index[:50])
plt.title('Number Of Terrorist Activities by Country (Top 50)')
plt.xticks(rotation=90)
plt.show()
```

```python
# number of incidents per region
plt.figure(figsize=(10,5))
# sns.countplot(data=terror,x='region')
sns.countplot(data=df,x='region_txt',palette='muted',
              order=df['region_txt'].value_counts().index)
plt.title('Number Of Terrorist Activities per Region')
plt.xticks(rotation=90)
```

```python
# Top 10 groups
top_10_gnames = df['gname'].value_counts().head(10)

# Hhorizontal bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_gnames.values, y=top_10_gnames.index)
plt.title('Top 10 Count of Terrorist Activities by Group')
plt.xlabel('Count')
plt.ylabel('Group Name')

plt.show()
```

```python
# Calculate the count of each weapon type
weapon_counts = df['weaptype1_txt'].value_counts()

# Select the top 10 most used weapons
top_10_weapons = weapon_counts.head(10)

# Create a new dataframe
top_10_table = pd.DataFrame({'Weapon Type': top_10_weapons.index, 'Count': top_10_weapons.values})

# Display the table
print(top_10_table)
```

```python
# Define the gnames and their corresponding top 3 weapons
gnames = ['Shining Path (SL)', 'Taliban', 'Farabundo Marti National Liberation Front (FMLN)']

# Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot the pie charts for each gname
for i in range(3):
    gname_data = df[df['gname'] == gnames[i]]
    print(f"Data for {gnames[i]}:")
    print(gname_data)
    if not gname_data.empty:
        top_3_weapons = gname_data['weaptype1_txt'].value_counts().head(3)
        axs[i].pie(top_3_weapons, labels=top_3_weapons.index, autopct='%1.1f%%', startangle=90)
        axs[i].set_title(gnames[i])

plt.suptitle('Top 3 Weapons by Selected gnames')
plt.show()
```

```python
# Filter the data for Iraq
iraq_data = df[df['country_txt'] == 'Iraq']

# Calculate the most present gname
most_present_gname = iraq_data['gname'].value_counts().head(5)

# Calculate the most used weapons in Iraq
most_used_weapons = iraq_data['weaptype1_txt'].value_counts().head(5)

# Create subplots for each information
fig, axs = plt.subplots(1,2, figsize=(15, 5))

# Plot the most present gname
axs[0].bar(most_present_gname.index, most_present_gname.values)
axs[0].set_title('Most Present gname in Iraq')
axs[0].set_xlabel('gname')
axs[0].set_ylabel('Count')
axs[0].tick_params(axis='x', rotation=90)

# Plot the most used weapons in Iraq
axs[1].bar(most_used_weapons.index, most_used_weapons.values)
axs[1].set_title('Most Used Weapons in Iraq')
axs[1].set_xlabel('Weapon')
axs[1].set_ylabel('Count')
axs[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Calculate the most attacked cities in Iraq
most_attacked_cities = iraq_data['city'].value_counts().head(10)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.barh(most_attacked_cities.index, most_attacked_cities.values)
plt.title('Top 10 Most Attacked Cities in Iraq')
plt.xlabel('Number of Attacks')
plt.ylabel('City')
plt.gca().invert_yaxis()  # Invert the y-axis to show cities from top to bottom

plt.show()
```

```python
# Calculate the main targeted group in Iraq
main_target = iraq_data['targtype1_txt'].value_counts().head(20)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.barh(main_target.index, main_target.values)
plt.title('Top 20 Most Targeted Group in Iraq')
plt.xlabel('Number of Attacks')
plt.ylabel('Target Group')
plt.gca().invert_yaxis()  # Invert the y-axis to show cities from top to bottom

plt.show()
```
